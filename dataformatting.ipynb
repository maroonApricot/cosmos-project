```python
import pandas as pd
```

```python
df = pd.read_csv("/data/class/cosmos2023/PUBLIC/shai_hulud/data/mirna/mirna_counts_matrix.csv")
metadata = pd.read_csv("/data/class/cosmos2023/PUBLIC/shai_hulud/data/mirna/mirna_metadata.csv")
```

```python
metadata.head()
```

```python
metadata_adj = metadata.replace('C2C12_NA','C2C12_0', regex=True)
metadata_adj = metadata_adj.replace('myotube_NA','myotube_72', regex=True)
for i in metadata_adj.index:
    if metadata_adj["Biosample.term.name"][i] == "C2C12":
        metadata_adj.at[i, "Age"] = metadata_adj.at[i, "age_long"] = 0
    elif metadata_adj["Biosample.term.name"][i] == "myotube":
        metadata_adj.at[i, "Age"] = metadata_adj.at[i, "age_long"] = 72
```

```python
metadata_adj["sample_mrna"]
```

```python
# Create a dictionary mapping Run to Sample from the metadata: 
column_mapping = dict(zip(metadata_adj['File.accession'], metadata_adj["sample_mrna"]))

# Rename the columns using the mapping
df.rename(columns=column_mapping, inplace=True)
```

```python
df.head()
```

```python
df = df.set_index(df['Unnamed: 0'])
df.index.names = ['gene_id']
df = df.drop('Unnamed: 0', axis=1)
df
```
