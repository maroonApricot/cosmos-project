```python
import pandas as pd
```

```python
df = pd.read_csv("data_adj/mirna_counts_matrix.csv")
df_cpm = pd.read_csv("data_adj/mirna_cpm_matrix.csv")
metadata = pd.read_csv("data_adj/mirna_metadata.csv")
```

```python
metadata_adj = metadata.replace('C2C12_NA','C2C12_0', regex=True)
metadata_adj = metadata_adj.replace('myotube_NA','myotube_72', regex=True)
for i in metadata_adj.index:
    if metadata_adj["Biosample.term.name"][i] == "C2C12":
        metadata_adj.at[i, "Age"] = metadata_adj.at[i, "age_long"] = 0
    elif metadata_adj["Biosample.term.name"][i] == "myotube":
        metadata_adj.at[i, "Age"] = metadata_adj.at[i, "age_long"] = 72
```

```python
metadata_adj.to_csv('data_adj/metadata.tsv', sep='\t')
```

```python
# Create a dictionary mapping Run to Sample from the metadata: 
column_mapping = dict(zip(metadata_adj['File.accession'], metadata_adj["Sample"]))

# Rename the columns using the mapping
df.rename(columns=column_mapping, inplace=True)
df_cpm.rename(columns=column_mapping, inplace=True)
```

```python
#counts
df = df.set_index(df['Unnamed: 0'])
df.index.names = ['gene_id']
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
#cpm
df_cpm = df_cpm.set_index(df_cpm['Unnamed: 0'])
df_cpm.index.names = ['gene_id']
df_cpm = df_cpm.drop('Unnamed: 0', axis=1)
df_cpm
```

```python
df = df.loc[df.sum(axis=1) != 0]
df_cpm = df_cpm.loc[df_cpm.sum(axis=1) != 0]
df.to_csv('data_adj/fixed_counts_matrix.tsv', sep='\t')
df_cpm.to_csv('data_adj/fixed_cpm_matrix.tsv', sep='\t')
```

## merging gene ref lists

```python
df_1 = pd.read_csv("data_adj/gene_to_mir.csv", usecols=['mirna_id', 'mirna_name'])
df_1.rename(
    columns={"mirna_id": "gene_id", "mirna_name": "gene_name"},
    inplace=True,
)
df_1 = df_1.loc[df_1["gene_name"].str.contains("Mir", case=False)].copy(deep=True)

df_2 = pd.read_csv("data_adj/transcripts_to_genes.txt", sep="\t", header=None, names=['transcript_id', 'gene_id', 'gene_name'])
df_2 = df_2.loc[df_2["gene_name"].str.contains("Mir", case=False)].copy(deep=True)
df_2 = df_2.drop(columns=['transcript_id'])
```

```python
merged = pd.concat([df_1, df_2])
merged.drop_duplicates(subset=['gene_name'], inplace=True)
```

```python
merged.to_csv("data_adj/mirna_id_to_name.tsv", sep="\t")
```
